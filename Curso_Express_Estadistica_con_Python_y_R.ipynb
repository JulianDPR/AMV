#### Arreglos, tuplas, diccionarios


```python
#---

#--Listas
#Operaciones
#append
#extend
#pop
#index
#+
#[::]
#[::-1]
#sort

A = ["1","2","3"]
A.append(2)
print(f"A: {A}")

B = A+["4"]
print(f"B: {B}")

C = A[::-1]
print(f"C: {C}")

ind = A.index(2)
print(f"Index (2): {ind}")

C = C.pop(ind)

print(f"A sin ind: {A}")

D = "+".join(C)
print(f"D: {D}")

A = [2,3,1]

A.sort()

ind = A.index(2)

print(f"A ordenada: {A}")

A[ind] = 0

print(f"A cambiando ({ind}): {A}")

#--Tuplas / Conjuntos
#Operaciones
#+
#[::]

A = (1,2,3)
print(f"Tupla A: {A}")
B = A+(4,5)
print(f"Tupla B: {B}")
C = A[::-1]
print(f"Tupla C: {C}")

#--Diccionarios
#Operaciones
#update
#del

A = {"B":1, "C":2}

A["D"] = 3

print(f"Diccionario A: {A}")

A.update({"E":4,"F":5,"G":6})

print(f"Diccionario A actualizado: {A}")

del A["B"]

print(f"Diccionario A sin B: {A}")


#---
```

#### Clases y funciones

```python
#--- Función simple

def suma(a,b):

  return(a+b)

print(suma(5,10))

#--- Función Rápida

a = lambda x: x[0]+x[1]

print(a([5,10]))

#--- Clase (Versión más potente)

class Math:

  def __init__(self,a,b):

    self.a = a
    self.b = b

  def suma(self):

    return self.a+self.b

  def resta(self):

    return self.b-self.a

  def mult(self):

    return self.b*self.a

Mat = Math(5,10)

print(Mat.suma())
```

#### Condicionales

```python
# Recordemos los booleanos, True y False

a = 5
b = 10
print(a < b)
print(a > b)
```

```python
# Ahora los condicionales inician con un "Si verdadero entonces haga:, sino haga:"

if (a < b):
  print("Si")
else:
  print("No")
```

```python
# También lo podemos plantear con su conjunto complemento p=>q, ~q=>~p

if not (a>b):
  print("Si")
else:
  print("No")
```

```python
#Por último la función if else

x = 5

print("Si") if x<5 else print("No")

#--- Los booleanos se pueden sumar True=1; False=0
print(True+False)
```

# Ciclos

```python
#--- For

A = [1,2,3,4,5,2]
a = 0
for i in A:

  a += i

print(a)
```

```python
#--- While

A = [1,2,3,4,5,2]
a = 0
while len(A) != 0:

  a += A[len(A)-1]
  A.pop()

print(a)
```

```python
#--- Mapeo

A = [1,2,3,4,5,2]
a = []
a.append(A[0])
#cumsum
b = list(map(lambda x: a.append(a[x-1]+A[x]), range(1,len(A))))
print(a)
print(a[-1])
```

#### Ejemplo de lo aprendido

```python
#--- Clases, Funciones y Ciclos

class Math:

  def __init__(self, A):

    self.A = A

  def suma(self):

    self.a = 0

    for i in range(len(self.A)):

      self.a += self.A[i]

    return(self.a)


Sum = Math(list(range(1,101)))

print(Sum.suma())

#n(n+1)/2 ; 100*(100+1)/2 = 5050
```

#### Vectorización

```python
# Vectorización de array

import numpy as np

# Recordemos las operaciones

A = [5,-1]
B = [3,7]

print(A+B)

# Vectorización

A = np.array(A)
B = np.array(B)

print(A+B)

# Multiplicación de Matrices

print(A@B)

# Redimensión

print(A.reshape(-1,1)@B.reshape(1,-1))

C=[[2,3],[4,5]]

# Inversa de una Matriz

print(np.linalg.inv(C))

```

##### Gráfico histograma

```python
# Gráfico histograma

F1.plot(kind="hist", rot=90)
```

```python
# Histograma seaborn

sns.histplot(F1, stat="proportion")
```

```python
# Histograma multiple

col = F1.shape[1]

fig, ax = plt.subplots(2,2)

palette = sns.color_palette("tab10",4)

ax = ax.ravel()

for i in range(col):

  sns.histplot(F1.iloc[:,i], ax=ax[i], stat="proportion", color=palette[i])

fig.tight_layout()
```

##### Gráfico de cajas y bigotes

```python
sns.boxplot(F1, palette=palette, orient="h")
```

##### Gráfico de calor

```python
sns.heatmap(F1.corr(),vmin=-1, vmax=1, cmap=sns.color_palette("vlag", as_cmap=True),annot=True)
```

##### Gráfico de dispersión

El mejor elemento para medir relaciones lineales entre dos variables

```python
sns.regplot(F1, x="sepal length (cm)", y="petal length (cm)", ci=95)
```

##### Pairplot

```python
pair = sns.pairplot(F1, diag_kind="kde")
pair.map_diag(sns.histplot, stat="density", color="skyblue")
pair.map_lower(sns.kdeplot, levels=4, color=".2")
pair.map_upper(sns.regplot)
```
