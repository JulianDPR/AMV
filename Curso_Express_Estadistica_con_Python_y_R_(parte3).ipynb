```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import prince as pc
import scipy.stats as ss
import altair as alt
alt.renderers.enable("colab")


F1 = pd.read_csv("/content/drive/MyDrive/Cursoexpress/LebartACS.csv", index_col=0)
print(F1)
n, p = F1.shape
```

### Prueba $\chi ^{2}$ de independencia

$H_0:Independencia \hspace{2cm} H_1:No\hspace{0.2cm}Independencia$

$Valor-p=P(\chi^{2}_{(n-1)(p-1)} \geq \chi^{2}_o)$

donde $\chi^{2}_o$ es el valor observado de la prueba que se calcula:

$\chi^{2}_o=\sum_{i=1}^{np}\frac{(O_i-E_i)^{2}}{E_i}$

```python
x2_o, pvalue, df, exp = ss.chi2_contingency(F1)
```

```python
F_1 = pc.CA(n_components=p).fit(F1)
```

#### Eig Summary

Se parte del mismo principio de ACP donde $\boldsymbol{S}$ es una matriz asímetrica conformada por.

$S = \frac{X'}{p*n}D_n^{-1}\frac{X}{p*n}D_p^{-1}$

Donde:

$U'D_p^{-1}SU=\lambda$

Y para las variables:

$T = \frac{X}{p*n}D_p^{-1}\frac{X'}{p*n}D_n^{-1}$

$V'D_n^{-1}TV=\lambda$

```python
F_1.eigenvalues_summary
```

```python
base = F_1.scree_plot().encode(x="eigenvalue",y="component")
base.mark_bar() + base.mark_text(align='left', dx=2)
```

#### Coordenadas de nuevos individuos

Para los individuos:

$\Psi=D_n^{-1}FD_p^{-1}U$

Para las variables:

$\Phi=D_p^{-1}F'D_n^{-1}V$

### Summary ACS

```python
def summary(CA_object, data, cat=None):

  import numpy as np
  import matplotlib.pyplot as plt
  import pandas as pd
  import seaborn as sns
  import altair as alt

  alt.renderers.enable("colab")

  n,p = data.shape

  summ = {}

  summ.update({"row_cor":CA_object.row_coordinates(data)})

  summ.update({"row_cont":CA_object.row_contributions_})

  summ.update({"row_cos":CA_object.row_cosine_similarities(data)})

  summ.update({"col_cor":CA_object.column_coordinates(data)})

  summ.update({"col_cont":CA_object.column_contributions_})

  summ.update({"col_cos":CA_object.column_cosine_similarities(data)})

  summ.update({"var":CA_object.eigenvalues_summary})

  base = CA_object.scree_plot().encode(x="eigenvalue",y="component")
  base.mark_bar() + base.mark_text(align='left', dx=2)

  summ.update({"plot_var":base})

  ind_var_graph1 = CA_object.plot(F1, x_component=0, y_component=1)

  ind_var_graph2 = CA_object.plot(F1, x_component=0, y_component=1).mark_text(align='left', dx=7).encode(text="value")

  ind_var_graph=ind_var_graph1+ind_var_graph2

  summ.update({"ind_var_graph":ind_var_graph})

  return summ

K = summary(F_1,F1)
#K["plot_var"]
#K["ind_var_graph"]
```
